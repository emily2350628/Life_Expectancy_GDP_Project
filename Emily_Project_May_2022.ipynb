```python
df.rename(columns = {'Life expectancy at birth (years)': 'Life_expectancy'}, inplace=True)
```

```python
avg_life_expectancies = []
for y in range(2000, 2016):
    #print('Average life expectancy in', y, ': ', round(np.mean(df[df.Year==y].Life_expectancy), 2))
    avg_life_expectancies.append(round(np.mean(df[df.Year==y].Life_expectancy), 2))
```

```python
avg_life_expectancies_zip = list(zip(range(2000, 2016), avg_life_expectancies))
```

```python
#avg_life_expectancies_zip
```

```python
plt.plot(range(2000, 2016), avg_life_expectancies)
plt.ylabel('Age')
plt.xlabel('Year')
plt.title('Life expectancy average of the six countries from 2000 to 2015')
```

```python
countries = ['Germany', 'Chile', 'Zimbabwe', 'China', 'Mexico', 'United States of America']
avg_life_expectancies_by_country = []
for c in countries:
    df_c = df[df.Country == c]
    for y in range(2000, 2016):
        #print('Average life expectancy in', c, ' in', y, ': ', round(np.mean(df_c[df_c.Year==y].Life_expectancy), 2))
        avg_life_expectancies_by_country.append(round(np.mean(df_c[df_c.Year==y].Life_expectancy), 2))
```

```python
countries_by_16 = []
for c in countries:
    for i in range(0, 16):
        countries_by_16.append(c)
#print(countries_by_16)
```

```python
avg_life_expectancies_by_country_zip = list(zip(avg_life_expectancies_by_country, countries_by_16))
```

```python
life_germany = avg_life_expectancies_by_country[0:16]
life_chile = avg_life_expectancies_by_country[16:32]
life_zimbabwe = avg_life_expectancies_by_country[32:48]
life_china = avg_life_expectancies_by_country[48:64]
life_mexico = avg_life_expectancies_by_country[64:80]
life_usa = avg_life_expectancies_by_country[80:96]
```

```python
life_by_country = [life_germany, life_chile, life_zimbabwe, life_china, life_mexico, life_usa]
```

```python
plt.plot(range(2000, 2016), life_germany, color = 'green')
plt.plot(range(2000, 2016), life_chile, color = 'red')
#plt.plot(range(2000, 2016), life_zimbabwe, color = 'yellow')
plt.plot(range(2000, 2016), life_china, color = 'purple')
plt.plot(range(2000, 2016), life_mexico, color = 'orange')
plt.plot(range(2000, 2016), life_usa, color = 'blue')
plt.legend(['Germany','Chile', 'China', 'Mexico', 'USA'])
plt.ylabel('Age')
plt.xlabel('Year')
plt.title('Life expectancy average of the six countries from 2000 to 2015')
```

```python
avg_gdp_yearly_all = []
for y in range(2000, 2016):
    #print('Average gdp in', y, ': ', round(np.mean(df[df.Year==y].GDP), 2))
    avg_gdp_yearly_all.append(round(np.mean(df[df.Year==y].GDP), 2))
```

```python
countries = ['Germany', 'Chile', 'Zimbabwe', 'China', 'Mexico', 'United States of America']
avg_gdp_yearly = []
for c in countries:
    df_c = df[df.Country == c]
    for y in range(2000, 2016):
        #print('Average life expectancy in', c, ' in', y, ': ', round(np.mean(df_c[df_c.Year==y].Life_expectancy), 2))
        avg_gdp_yearly.append(round(np.mean(df_c[df_c.Year==y].GDP), 2))
```

```python
plt.plot(range(2000, 2016), avg_gdp_yearly_all)
plt.title('average gdp total for all countries by year')
plt.ylabel('GDP(in trillions of USD)')
plt.xlabel('year')
```

```python
gdp_germany = avg_gdp_yearly[0:16]
gdp_chile = avg_gdp_yearly[16:32]
gdp_zimbabwe = avg_gdp_yearly[32:48]
gdp_china = avg_gdp_yearly[48:64]
gdp_mexico = avg_gdp_yearly[64:80]
gdp_usa = avg_gdp_yearly[80:96]
```

```python
plt.plot(range(2000, 2016), gdp_germany, color = 'green')
plt.plot(range(2000, 2016), gdp_chile, color = 'red')
plt.plot(range(2000, 2016), gdp_zimbabwe, color = 'yellow')
plt.plot(range(2000, 2016), gdp_china, color = 'purple')
plt.plot(range(2000, 2016), gdp_mexico, color = 'orange')
plt.plot(range(2000, 2016), gdp_usa, color = 'blue')
plt.legend(['Germany','Chile', 'Zimbabwe', 'China', 'Mexico', 'USA'])
plt.ylabel('GDP')
plt.xlabel('Year')
plt.title('GDP of the six countries from 2000 to 2015')
```
